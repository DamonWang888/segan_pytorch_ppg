```python
import numpy as np
import heartpy as hp
from scipy.io import loadmat
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20,20)
def bandpass(signal,sample_rate,frequency_field):
    pass_signal=hp.filtering.filter_signal(signal,cutoff=[0.1,5],sample_rate=sample_rate,order=3,filtertype='bandpass')
    return pass_signal

frequency_field=[0.1,5]
original_signal=loadmat('/home/wcj/CurrentProject/ppgalgorithm/Troika/Data/0716/02/DATA_S01_T01_E01generate.mat')
signal=original_signal['cleansig'][0]
plt.subplot(211)
plt.plot(signal[1:1000])
#print(signal.shape)
handled_signal=bandpass(signal,125,frequency_field)
plt.subplot(212)
plt.plot(handled_signal[1:1000])
plt.show()
```
